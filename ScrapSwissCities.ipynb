```python
# import libraries
from bs4 import BeautifulSoup
import urllib.request
import csv
import re
```

```python
rows = []
rows.append(['Canton', 'City', 'Longitude', 'Latitude'])
print(rows)

# specify the url
urlpage = 'https://fr.wikipedia.org/wiki/Liste_des_communes_de_Suisse'
# query the website and return the html to the variable 'page'
page = urllib.request.urlopen(urlpage)
# parse the html using beautiful soup and store in variable 'soup'
soup = BeautifulSoup(page, 'html.parser')
res = soup.find_all(href=re.compile("Communes"))
limit = 0
for link in res:
    # go to link 
    url = 'https://fr.wikipedia.org'+link.get('href')
    page2 = urllib.request.urlopen(url)
    
    # parse the html
    soup2 = BeautifulSoup(page2, 'html.parser')
    print('==',soup2.h1.getText())
    table = soup2.find('table').find_all('tr')
    for td in table:
        if (td.find('td')!=None):
            link2 = td.find('td').find('a')
            # jump to link of the city
            if (link2!=None):
                url2 = 'https://fr.wikipedia.org'+link2.get('href')
                page3 = urllib.request.urlopen(url2)
                soup3 = BeautifulSoup(page3, 'html.parser')
                res4 = soup3.find("a",{"class":"mw-kartographer-maplink"})
                print(soup3.h1.getText(),':',res4['data-lat'],',',res4['data-lon'])
                rows.append([soup2.h1.getText(),soup3.h1.getText(),res4['data-lat'],res4['data-lon']])
    
# Create csv and write rows to output file
with open('SwitzerlandCities.csv','w', newline='') as f_output:
    csv_output = csv.writer(f_output, delimiter=';')
    csv_output.writerows(rows)
  

    
```
