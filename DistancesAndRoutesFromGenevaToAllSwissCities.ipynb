```python
import csv
import math
from itertools import islice
import requests
import time

from routingpy import MapboxValhalla
from pprint import pprint
from routingpy import ORS
from routingpy.routers import options
```

```python
#Genève coordinates
nameVilleRef = 'Genève'
latitudeVilleRef = float(46.200013) 
longitudeVilleRef = float(6.149985)

# Create your own API key at https://openrouteservice.org/sign-up
api = ORS(api_key='openrouteservice_API_KEY')
options.default_skip_api_error = True

finalRows = []
finalRows.append(['Canton', 'City', 'Longitude', 'Latitude','TargetCity','DirectDistance','RouteDuration'])

with open('SwitzerlandCities.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=';', quotechar='"')
    next(reader, None)  # skip the headers
    for row in islice(reader,0,100):
        time.sleep(2)
        # Distance directe
        B2 = float(row[2])
        C2 = float(row[3])
        B3 = latitudeVilleRef
        C3 = longitudeVilleRef
        distDirecte = math.acos(math.sin(math.radians(B2))*math.sin(math.radians(B3))+math.cos(math.radians(B2))*math.cos(math.radians(B3))*math.cos(math.radians(C2-C3)))*6371
        #print(row[1],'=>Distance directe en km : ',distDirecte)
        
        # Distance trajet
        coordinates = [[longitudeVilleRef, latitudeVilleRef], [float(row[3]), float(row[2])]]
        try:
            route = api.directions(locations=coordinates, profile='driving-car')
            #print("Duration in minutes: {}".format(route.duration/60))
            routeDuration = format(format(route.duration/60))
        except TypeError:
            #print("Oops!  That was no valid number.  Try again...")
            routeDuration = 0
            
        #print(row[1],'=>Distance directe en km : ',distDirecte, ' Trajet en min : ',routeDuration)   
        print(row[0],',',row[1],',',row[2],',',row[3],nameVilleRef,' ',distDirecte,' ',routeDuration)
        finalRows.append([row[0],row[1],row[2],row[3],nameVilleRef,distDirecte,routeDuration])

        
# Create csv and write rows to output file
with open('SwitzerlandCitiesWithDistances.csv','w', newline='') as f_output:
    csv_output = csv.writer(f_output, delimiter=';')
    csv_output.writerows(finalRows)
    
print('Ended !')
```
